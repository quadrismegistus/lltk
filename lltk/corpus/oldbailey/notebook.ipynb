```python
lltk.barplot(C.meta, 'gender')
```

```python
tfidf = C.dtm(tfidf=True,meta=['gender'])
tfidf
```

```python
tfidf_gender=tfidf.groupby('gender').median().T
tfidf_gender
```

```python
import plotnine as p9
fig=p9.ggplot(tfidf_gender[['male','female']], p9.aes(x='female',y='male'))
fig+=p9.geom_point()
fig
```
